# Format Abilities
Takes in an ability name and description and outputs formatted text

```python
from __future__ import print_function, division

f = open('./template.txt','r')
lines = [x.strip() for x in f.readlines()]
f.close()
f = open('./output.txt','w')
for line in lines:
    ability, description = line.split('|')
    f.write('{\n')
    f.write(f'    "name":"{ability}",\n')
    f.write(f'    "description":[\n')
    cLine = ''
    words = description.split(' ')
    for word in words:
        if (len(cLine) + len(word) + 1) > 80:
            f.write(f'        "{cLine}",\n')
            cLine=''
        cLine += f'{word} '
    f.write(f'        "{cLine}"\n')
    f.write('    ]\n}\n')
f.close()
```

# Generator
Generates HTML from a template

```python
from __future__ import print_function, division
from collections import namedtuple
Section = namedtuple(
    'Section',
    ('type','content')
)

data = """
t|Test
p|This is a test paragraph
p|This is a second test paragraph
"""

lines = []
for line in [ 
    x.strip() 
    for x in data.split('\n') 
    if x.strip() != ''
]:
    comp = line.split('|')
    lines.append(Section(comp[0],comp[1]))

title = ''
for i in range(0, len(lines)-1):
    if lines[i].type == 't':
        title = lines[i].content
        lines.pop(i)
generated = "self.__add_section(\n"
generated+=f'    "{title}",\n    '
for line in lines:
    if line.type[0] == 'h':
        c = line.content
        if 'u' in line.type:
            c = f'<u>\n            {c}\n        </u>'
        generated += 'self.__htag(\n        """\n'
        generated += f'        {c}'
        generated += '\n        """\n    )+'
    if line.type == 'p':
        generated += 'self.__paragraph(\n        """\n'
        l = ''
        comp = line.content.split(' ')
        for c in comp:
            if (len(l) + len(c) + 1) > 80:
                generated += f'            {l}\n'
                l = ''
            l += f'{c} '
        generated += f'        {l}\n'
        generated += '        """\n    )+'
generated = f'{generated[:-1]}\n'
generated+= ")"

print(generated)
```

# Format Text
wraps up text and adds quotes and commas

```python
import re
inputs = [
"""
Summoning spells allow the caster to conjure powerful allies both to protect
themselves as well as to turn the tide of battle in cases where they themselves
or their party are overwhelmed. Summoned entities are controlled by the entity
who summoned them. In the event an NPC summons an entity; both the NPC and the
summoned entity are controlled by the DM. When a summoned entity dies; they
leave behind no corpse or items. Of course, as with all things in the system;
the DM has the power to change this.
"""
]

line = ""
for inp in inputs:
    line=""
    inp = re.sub(r'[\ \n]+',' ',inp)
    inp = inp.strip()
    for word in inp.split(' '):
        if (len(line) + len(word) + 1) > 80:
            print(f'"{line[:-1]}",')
            line=""
        line+=f'{word} '
    print(f'"{line[:-1]}"')
    print()
```

# Generate Spell json
Generates a json entry for a spell

```python
import re
def generate_text_block(ind_level, text):
    inp = re.sub(r'[\ \n]+',' ',text)
    line = ""
    indent = '    '*ind_level
    output = ""
    line=""
    inp = inp.strip()
    for word in inp.split(' '):
        if (len(line) + len(word) + 1) > 80:
            output += f'{indent}"{line[:-1]}",\n'
            line=""
        line+=f'{word} '
    output += f'{indent}"{line[:-1].strip()}"'
    return output

def print_spell(spell):
    ob='{'
    cb='}'
    indent='    '
    print('{')
    print(f'{indent}"name":"{spell["name"]}",')
    print(f'{indent}"description":[')
    desc = generate_text_block(2, spell['description'])
    if len(desc.strip(' "')):
        print(desc)
    print(f'{indent}],')
    print(f'{indent}"element":"{spell["element"]}",')
    print(f'{indent}"max-level":{spell["max-level"]},')
    print(f'{indent}"mp-cost-pl":{spell["mp-cost-pl"]},')
    print(f'{indent}"req-stat":[')

    # Print out required stats
    if len(spell['req-stat']):
        stat_str = ""
        for stat in spell['req-stat']:
            stat_str += f'{indent*2}"{stat}",\n'
        stat_str = f'{stat_str[:-2]}'
        print(stat_str)

    print(f'{indent}],')
    print(f'{indent}"effects":{ob}')
    print(f'{indent*2}"tags":[')

    # print out tags
    if len(spell['effects']['tags']):
        tag_str = ''
        for tag in spell['effects']['tags']:
            tag_str += f'{indent*3}{ob}'
            for key in tag:
                val = isinstance(tag[key],str) and f'"{tag[key]}"' or f'{tag[key]}'
                tag_str+= f'"{key}":{val}, '
            tag_str = f'{tag_str[:-2]}{cb},\n'
        tag_str = f'{tag_str[:-1]}'
        print(tag_str)

    print(f'{indent*2}],')
    print(f'{indent*2}"description":[')
    desc = generate_text_block(3, spell['effects']['description'])
    if len(desc.strip(' "')):
        print(desc)
    print(f'{indent*2}]')
    print(f'{indent}{cb},')
    print(f'{indent}"notes":[')
    desc = generate_text_block(2, spell['notes'])
    if len(desc.strip(' "')):
        print(desc)
    print(f'{indent}]')
    print('},')
    
print_spell({
    "name":"Summon Shogoth",
    "description":"""
    The caster calls out for help from an alternate dimension.
    """,
    "element":"dark",
    "max-level":2,
    "mp-cost-pl":4,
    "req-stat":['INT 1','INT 3'],
    "effects":{
        "tags":[
            {'effect':'whose horrible visage casts fear on target entity'},
            {'effect':'to fight alongside the user'}
        ],
        "description":"""
        User summons a shogoth from an alternate dimension {effect}
        """
    },
    "notes":"""
    Summoned entities drop nothing when killed. Entity is summoned adjacent to caster.
    Summoner controls actions for summoned entity. Summoned entity rolls initiative
    immediately when summoned.
    """
})
```
